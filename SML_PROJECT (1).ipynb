```python
import pandas as pd
data=pd.read_csv('/content/Advertising.csv')
data
```

```python
data.head()
```

```python
x=data[['TV','Radio','Newspaper']]
y=data[['Sales']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(X_train)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_model=LinearRegression()
linear_model.fit(X_train,y_train)
```

```python
y_pred_linear=linear_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mae = mean_absolute_error(y_test,y_pred_linear)
mae
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse=mean_squared_error(y_test,y_pred_linear)
mse
```

```python
import numpy as np
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error
mse_linear = mean_squared_error(y_test, y_pred_linear)
rmse = np.sqrt(mse_linear)
rmse
```

```python
from sklearn.svm import SVR
```

```python
svr_model = SVR(kernel='rbf')
svr_model.fit(X_train, y_train)
```

```python
y_pred_svr = svr_model.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, y_pred_svr)
mae
```

```python
mse = mean_squared_error(y_test, y_pred_svr)
mse
```

```python
rmse=np.sqrt(mse)
rmse
```

```python
from sklearn.linear_model import Lasso,Ridge
```

```python
Lasso_model=Lasso(alpha=1.0)
Lasso_model.fit(X_train,y_train)
```

```python
y_pred_lasso=Lasso_model.predict(X_test)
```

```python
mae=mean_absolute_error(y_test,y_pred_lasso)
mae
```

```python
mse=mean_squared_error(y_test,y_pred_lasso)
mse
```

```python
rmse=np.sqrt(mse)
rmse
```

```python
ridge_model=Ridge(alpha=1.0)
ridge_model.fit(X_train,y_train)
```

```python
y_pred_ridge=ridge_model.predict(X_test)
```

```python
mae=mean_absolute_error(y_test,y_pred_ridge)
mae
```

```python
mse=mean_squared_error(y_test,y_pred_ridge)
mse
```

```python
rmse=np.sqrt(mse)
rmse
```

```python

from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
mae_linear = mean_absolute_error(y_test, y_pred_linear)
mse_linear = mean_squared_error(y_test, y_pred_linear)
rmse_linear = np.sqrt(mse_linear)
result1 = {
    'Linear Regression': {'MAE': mae_linear, 'MSE': mse_linear, 'RMSE': rmse_linear}
}
result1
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
mae_svr = mean_absolute_error(y_test, y_pred_svr)
mse_svr = mean_squared_error(y_test, y_pred_svr)
rmse_svr = np.sqrt(mse_svr)
result2 = {
    'SVR': {'MAE': mae_svr, 'MSE': mse_svr, 'RMSE': rmse_svr}
}
result2
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
mae_lasso = mean_absolute_error(y_test, y_pred_lasso)
mse_lasso = mean_squared_error(y_test, y_pred_lasso)
rmse_lasso = np.sqrt(mse_lasso)
result3 = {
    'Lasso Regression': {'MAE': mae_lasso, 'MSE': mse_lasso, 'RMSE': rmse_lasso},
}
result3
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
mae_ridge = mean_absolute_error(y_test, y_pred_ridge)
mse_ridge = mean_squared_error(y_test, y_pred_ridge)
rmse_ridge = np.sqrt(mse_ridge)
result4 = {
    'Ridge Regression': {'MAE': mae_ridge, 'MSE': mse_ridge, 'RMSE': rmse_ridge}
}
result4
```
